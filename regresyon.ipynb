## Verimizin düzenleme işlemlerinin yapılması 

```python
energy.var()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
energy_scaler = pd.DataFrame(scaler.fit_transform(energy),columns=energy.columns)
```

```python
energy_scaler.head(5)
```

```python
energy_scaler.var()
```

## Verilerimizi eğitmek için ayrıma işlemini yapalım.

```python
X = energy_scaler.drop(['price actual'],axis=1).values
Y = energy_scaler['price actual'].values
```

```python
X
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(X,Y,test_size=0.10,random_state=42)
x_train,x_test,y_train,y_test=train_test_split(x_train,y_train,test_size=0.20,random_state=42)
```

## Linear Regresyon Modeli ile Eğitmek

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
linearRegresyon = LinearRegression()
linearRegresyon.fit(x_train,y_train)
y_pred=linearRegresyon.predict(x_val)
print('Train success rate :',linearRegresyon.score(x_train,y_train))
print('Validation success rate :',r2_score(y_val,y_pred))
```

## Ridge Regresyon Modeli İle Eğitmek

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.1,normalize=True)
ridge.fit(x_train,y_train)
y_pred=ridge.predict(x_val)
print('Train success rate :',ridge.score(x_train,y_train))
print('Validation success rate :',r2_score(y_val,y_pred))
```

## Polinom Regresyon Modeli İle Eğitmek

```python
from sklearn.preprocessing import PolynomialFeatures
obje_poly=PolynomialFeatures(degree=2)

x_train_poly=obje_poly.fit_transform(x_train)
x_val_poly=obje_poly.fit_transform(x_val)
x_test_poly=obje_poly.fit_transform(x_test)

model_poly=LinearRegression(n_jobs=-1).fit(x_train_poly,y_train)
y_pred=model_poly.predict(x_val_poly)

print('Train success rate :',model_poly.score(x_train_poly,y_train))
print('Validation success rate :',r2_score(y_val,y_pred))
```

## Rastgele Orman Regresyon Modeli İle Eğitmek

```python
from sklearn.ensemble import RandomForestRegressor
model_rfr=RandomForestRegressor(n_estimators=100,n_jobs=-1,random_state=42).fit(x_train,y_train)
y_pred=model_rfr.predict(x_val)
print('Train success rate :',model_rfr.score(x_train,y_train))
print('Validation success rate :',r2_score(y_val,y_pred))
```

## AdaBoost Regresyon Modeli İle Eğitmek

```python
from sklearn.ensemble import AdaBoostRegressor
model_ada=AdaBoostRegressor(n_estimators=10,random_state=42).fit(x_train,y_train)
y_pred=model_ada.predict(x_val) 
print('Train success rate : ',model_ada.score(x_train,y_train))
print('Validation success rate : ',r2_score(y_val,y_pred))
```
